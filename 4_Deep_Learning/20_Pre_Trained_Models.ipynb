```python
# Import Library
from keras.applications import VGG16
from keras.applications.vgg16 import decode_predictions, preprocess_input
from tensorflow.keras.preprocessing import image
import numpy as np
```

```python
# Load The VGG16 Model and Specify Input Shape
model = VGG16(weights='imagenet', include_top=True, input_shape=(224, 224, 3))
```

```python
# Load an Image and Preprocess it for the Model
img_path = r"D:\COURSE\DATA SCIENCE FULL COURSE BY UDEMY\SECTION 8 Deep Learning\Deep Learning - Resources-20260311T130352Z-3-001\Deep Learning - Resources\LSTM\microphone.jpg"
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
# Use Model To Predict The Class
pred = model.predict(x)
```

```python
# Get The Top 5 Predictions with Class Names
decode_preds = decode_predictions(pred, top=5)[0]

for pred in decode_preds:
    print(pred[1], ':', pred[2])
```
