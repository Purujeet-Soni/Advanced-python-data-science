# Library Import

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as st

# Sciket Learn
from sklearn.linear_model import LogisticRegression as lr, LinearRegression 
from sklearn.preprocessing import StandardScaler as ss, OneHotEncoder as ohe
from sklearn.metrics import classification_report as cr, confusion_matrix as cm, accuracy_score as AS, recall_score as rc, precision_score as ps, f1_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split as tts, GridSearchCV as gscv
from sklearn.ensemble import RandomForestClassifier as rfc
from sklearn.tree import DecisionTreeClassifier as dtc, plot_tree as pt, export_graphviz
from sklearn.neighbors import KNeighborsClassifier as knc
from sklearn.svm import SVC
from sklearn.compose import ColumnTransformer as ct
from sklearn.naive_bayes import MultinomialNB as nb
from sklearn.naive_bayes import GaussianNB as gnb
```

# Data Import

```python
# Data Importing
dat = pd.read_csv(r"E:\DATA FOR TEST\Tele Customer churn\WA_Fn-UseC_-Telco-Customer-Churn.csv")
dat['TotalCharges'] = pd.to_numeric(dat.TotalCharges, errors = 'coerce')
data = dat.copy()
data['TotalCharges'] = data['TotalCharges'].fillna(data['TotalCharges'].mean())
data.head()
```

# EDA

```python

print(data.info())
print(data.describe())
print(data.isnull().sum())

# Drop ID column (FIXED)
data.drop(columns=['customerID'], inplace=True)

# Split Data in X and Y
x = data.iloc[:, :-1]
y = data.iloc[:, -1]

# Train-test split
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.2, random_state = 42)

# Feature Encoding One HOT Encoder
cat_col = x_train.select_dtypes(include=['object', 'category']).columns

one_hot = ct(transformers = [('encoder', ohe(handle_unknown = 'ignore'), cat_col)], remainder = 'passthrough')

x_train_encoded = one_hot.fit_transform(x_train)
x_test_encoded = one_hot.transform(x_test)

# Feature Scaling
scaler = ss()
x_train = scaler.fit_transform(x_train_encoded)
x_test = scaler.transform(x_test_encoded)
```

# Machine Learning Model One by One

```python
# Machine Learning logistic Model Initilization
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
y_pred_logistic_model = logistic_model.predict(x_test)
print("Classification Report :\n", cr(y_test, y_pred_logistic_model))

c_matrics = cm(y_test, y_pred_logistic_model)
print("Confusion Matrix logistic_model:\n", c_matrics)

TN, FP, FN, TP = c_matrics.ravel()
total = TN + FP + FN + TP

print("\nDetails (Counts and Percentages):")
print(f"True Negative (TN): {TN} ({TN/total*100:.2f}%)")
print(f"True Positive (TP): {TP} ({TP/total*100:.2f}%)")
print(f"False Positive (FP): {FP} ({FP/total*100:.2f}%)")
print(f"False Negative (FN): {FN} ({FN/total*100:.2f}%)")
```

# One Function For all Classification Data

```python
def find_best_model(x, y):

    models = {
        "LogisticRegression": lr(max_iter=1000),
        "RandomForest": rfc(),
        "DecisionTree": dtc(),
        "KNN": knc(),
        "SVC": SVC()
    }

    x_train, x_test, y_train, y_test = tts(
        x, y, test_size=0.2, random_state=42,
        shuffle=True,
        stratify=y
    )

    cat_cols = x.select_dtypes(include=['object', 'category']).columns
    num_cols = x.select_dtypes(include=['int64', 'float64']).columns

    preprocessor = ct(
        transformers=[
            ('encoder', ohe(handle_unknown='ignore'), cat_cols),
            ('scaler', ss(), num_cols)
        ]
    )

    for name, model in models.items():
        print(f"\n===== {name} =====")
        pipe = Pipeline([
            ('prep', preprocessor),
            ('model', model)
        ])

        pipe.fit(x_train, y_train)
        y_pred = pipe.predict(x_test)

        print(cr(y_test, y_pred))

        cmatrix = cm(y_test, y_pred)
        print("Confusion Matrix:\n", cmatrix)
        
        TN, FP, FN, TP = cmatrix.ravel()
        total = TN + FP + FN + TP

        print(f"TN: {TN} ({TN/total*100:.2f}%)")
        print(f"FP: {FP} ({FP/total*100:.2f}%)")
        print(f"FN: {FN} ({FN/total*100:.2f}%)")
        print(f"TP: {TP} ({TP/total*100:.2f}%)")

find_best_model(x, y)
```
