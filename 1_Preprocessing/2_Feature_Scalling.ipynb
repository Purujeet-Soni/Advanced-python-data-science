# Outlier Treatment

```python
def find_Outliers(df):
    outliers = []
    data_std_dev = df.std()
    data_Mean = df.mean()
    lower_limit = data_Mean - data_std_dev * 3
    upper_limit = data_Mean + data_std_dev * 3
    for Outlier in df:
        if Outlier > upper_limit or Outlier < lower_limit:
            outliers.append(Outlier)
    return outliers
```
